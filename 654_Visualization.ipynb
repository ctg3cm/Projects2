ii. Why use a coefficient of variation? How does it offer us insight that we cannot always get from a standard deviation alone, when we want to compare the dispersion between two variables?

Both the standard deviation and the coefficient of variation measure the spread of values in a dataset. The standard deviation measures how far the average value lies from the mean. The coefficient of variation measures the ratio of the standard deviation to the mean. The higher the coefficient of variation, the higher the standard deviation of a sample relative to the mean.

The standard deviation is most commonly used when we want to know the spread of values in a single dataset.

However, the coefficient of variation is more commonly used when we want to compare the variation between two datasets.

In this case, the standard deviation tells us that the typical value in this dataset lies 15407 and 10422 units away from the mean. The coefficient of variation then tells us that the standard deviation is 0.33 and 0.43 of the sample mean.

Reference:

Zach. (2021, May 17). Coefficient of variation vs. standard deviation: The difference. Statology. Retrieved September 20, 2022, from https://www.statology.org/coefficient-of-variation-vs-standard-deviation/
