```python
import sqlite3
from sqlite3 import Error
import pandas as pd

# connecting to SQL
connection = sqlite3.connect("chinook.db")

# SQL query to read data from a database
def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None 
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as e:
        print(f"The Error '{e}' occurred.")

# returned query results using fetchall()
q3 = '''
        SELECT i.BillingCountry, ROUND(SUM(total), 2) 
        AS 'TotalByCountry' 
        FROM invoices AS i
        GROUP BY BillingCountry
        ORDER BY TotalByCountry DESC;       
    ''' 

out = execute_read_query(connection, q3)

# SQL resultes as a Python DataFrame
name_df = pd.DataFrame(data = out, 
                       columns = ['BillingCountry','TotalByCountry'])
name_df[:]
```
