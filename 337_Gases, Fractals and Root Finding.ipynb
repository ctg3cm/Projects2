### Ideal gas law

The ideal gas law, also called the general gas equation, is the equation of state of a hypothetical ideal gas. It is a good approximation of the behavior of many gases under many conditions, although it has several limitations. (Wiki) The equation of the ideal gas is: 

$$PV=nRT.$$ 

$P, V$ and $T$ are the pressure, volume of container and the temperature of the gas respectively in this equation. $n$ is the number of models of gas. $R$ is the universal gas constant, which in SI units is $8.314 J/mol/K$.

### Van der Waals equation

The van der Waals equation (or van der Waals equation of state; named after Johannes Diderik van der Waals) is based on plausible reasons that real gases do not follow the ideal gas law. To account for the volume that a real gas molecule takes up, the van der Waals equation replaces $V$ in the ideal gas law with $(V-b)$, where b is the volume per mole that is occupied by the molecules. (Wiki) The equation is:

$$(P + \frac{an^2}{V^2})(V-b)=nRT.$$

### The reduced form

The van der Waals equation can also be expressed in terms of reduced properties. This yields a critical compressibility factor of 3/8. The equation is: 

$$(p_R + \frac{3}{V_R^2})(v_R - \frac{1}{3}) = \frac{8}{3}T_R.$$

### Critical point of van der Waals equation. 

In thermodynamics, a critical point is the end point of a phase equilibrium curve.(Wiki)

When $$T > T_C$$ The gas canoot be liquefy. 

However, when $$T < T_C$$ The gas can be liquefy when the pressure goes up, and the prssure will go down when the temperature goes down. 

The reduced form can be expanded to:
$$v_R^3 - \frac{1}{3}(1 + \frac{8T_R}{p_R})v_R^2 + \frac{3}{p_R}v_R - \frac{1}{p_R} = 0.$$

### Bisection method
The bisection method in mathematics is a root-finding method that repeatedly bisects an interval and then selects a subinterval in which a root must lie for further processing. It is a very simple and robust method, but it is also relatively slow. Because of this, it is often used to obtain a rough approximation to a solution which is then used as a starting point for more rapidly converging methods.(Wiki)

The algorithm is as follows:

 1. Choose a "bracket" $[a, b]$ of the root such that $sign(f(a)) \neq sign(f(b))$.
 2. Calculate $f$ at the midpoint of the bracket i.e. $c = (a + b)/2$.
 3. Replace $a$ or $b$ in such a way that the *new* $f(a)$ and $(f(b)$ also have opposite signs.
 4. Keep repeating steps 1-3 till a certain tolerance is reached or if the bracket width $|b - a|$ reaches a certain defined minimum.
 5. The final reported answer is the last value of $c$ which meets tolerance.(MTH337 Week7's note)

```python
def myFunction(x = 0.1):
    """
    Return the value os cos(x) - x. x is in radians
    
    Keywords: 
    
    x - input for the function - must be in radians
    
    """
    return np.cos(x) - x
```

```python
def BisectionFunc(f = myFunction, a= 0.4, b = 1.0, tolerance=1e-7):
    """
    Returns the root of the input function and number of iterations
    as a pair of values using the bisection method
    
    Keywords:
    f - function whose root I want to calculate
    a = start of the interval
    b - end of the interval
    tolerance - number of sig figs I need
    
    """
    
    signa = np.sign(f(a))
    signb = np.sign(f(b))
    numIter = 0
    
    while (b-a > tolerance):
        c = (a+b)/2
        print("Best Guess = {}".format(c))
        signc = np.sign(f(c))
        
        # What happens if signc == 0????
        
        if signc == signa:
            a = c
        else: 
            b = c
        numIter +=1
        
    return c, numIter
```

### Newton's method

The newton's method is a method for finding successively better approximations to the roots of a real-valued function. (Wiki)

 Basically, the algorithm works as follows

 1. The function must be differentiable (note that Bisection does not require this).
 2. First find the tangent to the function $f(x)$ at the current position $x_i$. Finding tangent means essentially calculating the derivative of the function.
 3. Extend the tangent line and calculate where it cuts the x-axis. That gives $x_{i+1}$ i.e. a new estimate of the root.
 4. Repeat till a certain tolerance is reached.(MTH337 Week7 note)

The iteration formula is:
 
 $$x_{i+1} = x_i - \frac{f(x_i)}{f'(x_i)}.$$
 

```python
def myFunction(x = 0.1):
    """Function returns the output cos(x) - x and its derivative 
       -sin(x) - 1
       
       Keywords:
       x - input
       output - cos(x) - x, -sin(x) - 1
    
    """
    
    return np.cos(x) - x, -np.sin(x) - 1
```

```python
def newtonMethod(fx = myFunction, xInit = 0.5, 
                 tolerance = 1e-8):
    """
    Function returns the root of the input function f and 
    the number of iterations it takes using Newton's method
    
    Keywords:
    
    f - function which return, f and df as a pair of values
    xInit - initial guess
    tolerance - number of sig figs I want to know reliably
    
    output - pair of values, first is the root, second 
             is the number of iterations
    """
    
    # Initialize the variables
    
    step = 0
    f, df = fx(xInit)
    dx = -f/df
    x = xInit
    
    # Run the iterations in the algorithm
    while np.abs(dx) > tolerance:
            print("Best guess of root = {} at iteration {:3d}".format(x, step))
            f, df = fx(x)
            dx = -f/df
            x += dx
            step += 1 # Just in case algorithm does not converge, stop 
                        # after some number of iterations.
                        # Build a check for this.
            
    return x, step
    
    
```
